```python
# python program to print the student mark sheet

subjects = ["MATHS", "SCIENCE", "HISTORY", "ENGLISH", "MARATHI"]
marks = []
count = 0

#student details 
name = input("Enter Your Name ")
std = int(input("Enter Your Standard: "))
div = input("Enter Your Division: ")
roll_no = input("Enter Your Roll Number ")
college = input("Enter Your College Name ")
 
    
# taking marks from user    
for x in subjects:
    a = int(input(f"Enter marks of {x}: "))
    marks.append(a)

    
# percentage calculation
total = sum(marks)
percentage = (total/500)*100

# grading system
if (percentage >= 75):
    class1 = "DISTINCTION" 
elif (percentage >=60 and percentage <= 74):
    class1 = "FIRST DIISION"
elif (percentage >= 45 and percentage <= 59):
    class1 = "SECOND DIVISION"
elif (percentage >= 35 and percentage <= 44):
    class1 = "PASS GRADE"
else:
    class1 = "FAILED"

if percentage < 35:
    count = count + 1
    
    
print("\n\n\t\t********************** STUDENT MARKSHEET **********************\n\n")
print("\t\tCOLLEGE: ", college.upper())
print("\n\t\tNAME: ", name.upper())
print("\n\t\tSTANDARD: ", std, "         ", "DIVISON: ", div)
print("\n\t\tROLL NUMBER: ",roll_no)
print("\n\t\tSUBJECTS                MARKS")
for (x, y) in zip(subjects, marks):
    print("\n\t\t", x, "             ","\t", y)
    
    
if count == 0:
    print("\n\t\tTOTAL MARKS: ", total, "\t\tRESULT:  PASS")
else:
    print("\n\t\tTOTAL MARKS:", total, "\t\tRESULT:  FAIL")
    
print("\n\t\tPERCENTAGE: ", percentage, "\t\tCLASS: ", class1)
#V.G. Vaze College Of Arts, Science and Commerce
```
